```python
def loadData():

  # df = pd.read_csv("gdrive/My Drive/Thesis/dataset/labeled_data.csv")
  # df = pd.read_csv("gdrive/My Drive/Thesis/dataset/train.csv")
  df = pd.read_csv("gdrive/My Drive/Thesis/dataset/tweeteval.csv")
  
  df = df[df['class'] == 0]
  df['class'].value_counts()
  df  = df[["tweet"]]

  def clean_text(txt):
    txt = re.sub(r'@[A-Za-z0-9]+', '', txt)
    txt = re.sub(r'#', '', txt)
    txt = re.sub(r'RT[\s]+', '', txt)
    txt = re.sub(r'https?:\/\/\S+', '', txt)
    txt = re.sub(r'\n', '', txt)
    filtered_sentence = remove_stopwords(txt)

    return filtered_sentence

  df['tweet'] = df['tweet'].apply(clean_text)

  # df.head()
  data = df.tweet.values.tolist()

  # return data[0:1000]
  return data,df
```

```python
data,df = loadData()
# df['class'].value_counts()
df.head()
topic_model = BERTopic(language="english")
topics, probs = topic_model.fit_transform(data)
```
